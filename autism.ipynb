```python
import matplotlib.pyplot as plt

# 1. Distribution of Result
# -------------------------------
plt.figure()
plt.hist(df['result'])
plt.title("Distribution of Result Score")
plt.xlabel("Result")
plt.ylabel("Frequency")
plt.show()


# -------------------------------
# 2. Result vs Numerical Features
# -------------------------------
numerical_features = ['age']

for feature in numerical_features:
    plt.figure()
    plt.scatter(df[feature], df['result'])
    plt.xlabel(feature)
    plt.ylabel('result')
    plt.title(f"Result vs {feature}")
    plt.show()


# -------------------------------
# 3. Result vs Categorical Features
# -------------------------------
categorical_features = [
    'gender', 'jaundice', 'austim',
    'used_app_before', 'relation'
]

for feature in categorical_features:
    plt.figure()
    df.boxplot(column='result', by=feature)
    plt.title(f"Result vs {feature}")
    plt.suptitle("")
    plt.show()


# -------------------------------
# 4. Correlation Heatmap
# -------------------------------
# Convert categorical to numeric
df_encoded = df.copy()
df_encoded = pd.get_dummies(df_encoded, drop_first=True)

plt.figure()
corr = df_encoded.corr()

plt.imshow(corr)
plt.colorbar()
plt.title("Correlation Heatmap")

plt.xticks(range(len(corr.columns)), corr.columns, rotation=90)
plt.yticks(range(len(corr.columns)), corr.columns)

plt.show()
```

```python
import seaborn as sns

sns.set_style("whitegrid")
sns.set_context("talk")

# -----------------------------------
# 1. Distribution of Result (Better)
# -----------------------------------
plt.figure()
sns.histplot(df['result'], kde=True)
plt.title("Distribution of Result Score")
plt.xlabel("Result")
plt.ylabel("Count")
plt.show()


# -----------------------------------
# 2. Age vs Result (BEST VERSION)
# -----------------------------------
plt.figure()
sns.regplot(x='age', y='result', data=df)  # regression + scatter
plt.title("Age vs Result (with Trend Line)")
plt.show()


# BONUS: Density-based (even better insight)
plt.figure()
sns.kdeplot(x=df['age'], y=df['result'], fill=True)
plt.title("Age vs Result Density Plot")
plt.show()


# -----------------------------------
# 3. Categorical vs Result (Premium Boxplots)
# -----------------------------------
categorical_features = [
    'gender', 'jaundice', 'austim',
    'used_app_before', 'relation'
]

for feature in categorical_features:
    plt.figure()
    sns.boxplot(x=feature, y='result', data=df)
    sns.stripplot(x=feature, y='result', data=df, alpha=0.3)  # adds points
    plt.title(f"{feature} vs Result")
    plt.show()


# -----------------------------------
# 4. Correlation Heatmap (Beautiful)
# -----------------------------------
df_encoded = pd.get_dummies(df, drop_first=True)

plt.figure(figsize=(12,8))
sns.heatmap(df_encoded.corr(), annot=False, cmap='coolwarm')
plt.title("Feature Correlation Heatmap")
plt.show()


# -----------------------------------
# 5. Result vs ASD Class (VERY IMPORTANT)
# -----------------------------------
plt.figure()
sns.boxplot(x='Class/ASD', y='result', data=df)
sns.stripplot(x='Class/ASD', y='result', data=df, alpha=0.3)
plt.title("Result vs ASD Class")
plt.show()
```
